```python
import subprocess
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import pandas
import time

```

```python


plt.rcParams["figure.figsize"] = (6,6)
algs = ('                     curl','                       wget', 'Min Energy D-Tree','Di Tacchio', 'Max Tput D-Tree')
y_pos = np.arange(len(algs))

print("y_pos = ",y_pos)

throughput_html = [0.492165203,0.499706159,813.222346,376.5803036,1057.545398] #intercloud

throughput_image = [8.907626522,6.134046375,1041.463453,934.4679256,1143.647454] #intercloud

throughput_video = [66.15395523,57.50305339,2187.521526,1208.915057,4877.799266] #Chameleon

#tput_mixed = [

#]

plt.bar(y_pos-0.3, throughput_html, align='center', width=0.2, color='none', edgecolor='red', hatch='xxx', lw=1, zorder=0, label='small files')
plt.bar(y_pos-0.1, throughput_image, align='center', width=0.2, color='none', edgecolor='purple', hatch='---', lw=1, zorder=0, label='medium files')
plt.bar(y_pos+0.1, throughput_video, align='center', width=0.2, color='none', edgecolor='blue', hatch='///', lw=1, zorder=0, label='large files')
#plt.bar(y_pos+0.3, tput_mixed, align='center', width=0.2, color='none', edgecolor='green', hatch='+++', lw=1, zorder=0, label='mixed')

plt.xticks(y_pos-0.4, algs,fontsize=16)
plt.xticks(rotation=50)
plt.ylabel('Avg. Throughput (Mbps)',fontsize=16)
plt.tight_layout()
# plt.legend(loc='upper left', frameon=False,fontsize=14)
plt.legend(loc='upper center',frameon=False,fontsize=16,bbox_to_anchor=(1.20, 1),
          ncol=1,fancybox=True, shadow=True)
plt.grid(axis = 'y')
#plt.yscale("log")
#plt.title('Chameleon Testbed (Client)')
#plt.savefig('/Users/usr/Downloads/Chameleon_Client_Throughput.pdf', format='pdf', dpi=1000)
plt.savefig('figures/intercloud_throughput.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.savefig('figures/intercloud_throughput.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show();
print(plt.rcParams.get('figure.figsize'))


plt.show();




```
